```python
from moviepy.editor import *
```

```python
speaker = 's25'
videoName= 'test_bgbszp'
```

```python
video= 'C:/Users/sally/Desktop/anaysis_result/individual_speaker/video/'+speaker+'/'+videoName+'.mpg'
outVideo ='C:/Users/sally/Desktop/anaysis_result/individual_speaker/video/'+speaker+'/test_silent.mp4'
```

### 1. Load original vedio and show the video

```python
videoclip = VideoFileClip(video)
# show original video clip
videoclip.ipython_display()
```

### 2. Remove audio and get silent video

```python
videoclip = VideoFileClip(video)
new_clip = videoclip.without_audio()
new_clip.write_videofile(outVideo)
# showing silent video clip
new_clip.ipython_display()
```

```python
# loading video dsa gfg intro video
clip = VideoFileClip(outVideo)
  
# getting only first 2.98 seconds
clip = clip.subclip(0, 2.98)
```

### 3. Add CNN audio into the silent video

```python
cnnAudio = 'C:/Users/sally/Desktop/anaysis_result/individual_speaker/audio/'+speaker+'/cnn_test.wav'
outCNN = 'C:/Users/sally/Desktop/anaysis_result/individual_speaker/video/'+speaker+'/test_cnn.mp4'


# loading audio file
cnnclip = AudioFileClip(cnnAudio).subclip(0, 2.98)
  
# adding audio to the video clip
cnnvideo = clip.set_audio(cnnclip)
cnnvideo.write_videofile(outCNN)

# showing cnn video clip
cnnvideo.ipython_display()
```

### 3. Add gp audio into the silent video

```python
gpAudio = 'C:/Users/sally/Desktop/anaysis_result/individual_speaker/audio/'+speaker+'/gp_test.wav'
outGP = 'C:/Users/sally/Desktop/anaysis_result/individual_speaker/video/'+speaker+'/test_gp.mp4'


# loading audio file
gpclip = AudioFileClip(gpAudio).subclip(0, 2.98)
  
# adding audio to the video clip
gpvideo = clip.set_audio(gpclip)
gpvideo.write_videofile(outGP)

# showing cnn video clip
gpvideo.ipython_display()
```
